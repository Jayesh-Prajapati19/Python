```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,8])
y=np.array([3,10])
plt.plot(x,y)
plt.show()
plt.plot(x,y,'o')
plt.show()
plt.plot(x,y,marker='o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])
x2=np.array([0,1,2,3])
y2=np.array([6,2,7,1])
plt.plot(x1,y1,x2,y2)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph",fontdict={'family':'serif','color':'blue','size':20},loc="left")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10*np.random.randint(100,size=(100))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
#bar graph
import matplotlib.pyplot as plt
import numpy as np
x=['A','B','C']
y=[400,350,500]
plt.bar(x,y,color="red",width=0.5)
plt.show()
plt.barh(x,y,color="black",height=0.5)
plt.show()
```

```python
#histogram- It gives the frequency of the element.
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randint(0,100,12)
plt.hist(x)
# plt.hist(x,bins=3)
# plt.hist(x,bins=[0,20,40,60,80,100],orientation='vertical')
plt.show()
```

```python
#Pie chart
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15,20])
# plt.pie(y,labels=["A","B","C","D","E"])
plt.pie(y,labels=["A","B","C","D","E"],explode=[0.1,0,0,0.1,0.5])
plt.legend(title="Four Colors",loc=0)
plt.show()
```

```python
#Sub Plot
y=np.array([3,8,1,10])
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.plot(y)
y=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(y)
plt.show()
```
