```python
import matplotlib.pyplot as plt
import numpy as np

# Data for Subplot 1
x1 = [5, 10, 25, 60, 80]
y1 = [5, 17, 25, 40, 30]

# Data for Subplot 2
m = [22, 30, 35, 35, 26]
w = [25, 32, 30, 35, 29]
bar_width=0.3
pos_men=range(len(m))
# print(pos_men)
pos_women=np.array(pos_men)+bar_width
# print(pos_women)


# Data for Subplot 3
car_companies = ['Maruti Suzuki', 'Hyundai', 'Kia', 'Toyota', 'Honda']
popularity = [25, 50, 30, 20, 35]

# Data for Subplot 4
math_marks = [88, 92, 80, 89, 100, 80, 60, 100, 80, 34]
science_marks = [35, 79, 79, 48, 100, 88, 32, 45, 20, 30]

# Data for Subplot 5
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C', 'C++']
popularity_languages = [20, 100, 25, 30, 45, 50]
c=["r","g","b","c","k","y"]

# Data for Subplot 6
h = [10, 20, 20, 30, 30, 30, 40, 40, 40, 40, 50, 50, 50, 60, 60, 70]

plt.figure(figsize=(20,20))
plt.subplot(3,2,1)
plt.plot(x1,y1,linestyle="dashed",marker="D",color="green")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Graph-1")

plt.subplot(3,2,2)
plt.bar(pos_men,m,color="y",width=0.3,label="Men")
plt.bar(pos_women,w,color="g",width=0.3,label="Women")
plt.legend()
plt.title("Graph-2")

plt.subplot(3,2,3)
plt.pie(popularity,labels=car_companies)
plt.title("Graph-3")

plt.subplot(3,2,4)
plt.scatter(np.arange(1,11),math_marks,color="blue",marker="o",label="Maths")
plt.scatter(np.arange(1,11),science_marks,color="red",marker="*",label="Science")
plt.title("Graph-4")

plt.subplot(3,2,5)
plt.barh(languages,popularity_languages,color=c)
plt.xlabel("Popularity")
plt.ylabel("Languages")
plt.title("Graph-5")

plt.subplot(3,2,6)
plt.hist(h,color="red")
plt.title("Graph-6")

plt.show()
```
